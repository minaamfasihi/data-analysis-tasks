```
import pandas as pd
import io

from google.colab import files
uploaded = files.upload()

df_c1 = pd.read_csv(io.StringIO(uploaded['code1.csv'].decode('utf-8')))
col_names = []
for col in df_c1.columns:
  col_names.append('file1_' + col)
df_c1.columns = col_names
df_c1
```

```
from google.colab import files
uploaded = files.upload()
df_c2 = pd.read_csv(io.StringIO(uploaded['code2.csv'].decode('utf-8')))
col_names = []
for col in df_c2.columns:
  col_names.append('file2_' + col)
df_c2.columns = col_names
df_c2
```

```
from google.colab import files
uploaded = files.upload()
df_p = pd.read_csv(io.StringIO(uploaded['products.csv'].decode('cp1252')))
col_names = []
for col in df_p.columns:
  col_names.append('file3_' + col)
df_p.columns = col_names
df_p
```

```
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

# def similar(str1, str2):
#     str1 = set(str1.split())
#     str2 = set(str2.split())
#     return float(len(str1 & str2)) / len(str1 | str2)

threshold = 0.2
df_ = pd.DataFrame(columns = ['file3_id', 'file3_description', 'status', 'file1_id', 'file1_segment1', 'file1_segment2', 'file1_segment3', 'file1_segment4', 'file1_code', 'file1_description', 'file1_probability', 'file2_id', 'file2_segment1', 'file2_segment2', 'file2_segment3', 'file2_segment4', 'file2_code', 'file2_description', 'file2_probability'])
for index, row in df_p.iterrows():
  highest_match_prob = 0
  highest_match_code = ""
  matching_id = ""
  segment1 = ""
  segment2 = ""
  segment3 = ""
  segment4 = ""
  description = ""
  for c1_ind, c1_row in df_c1.iterrows():
    result = similar(row.file3_description, c1_row.file1_description)
    if highest_match_prob < result:
      highest_match_prob = result
      highest_match_code = c1_row.file1_code
      description = c1_row.file1_description
      matching_id = c1_row.file1_id
      segment1 = c1_row.file1_segment1
      segment2 = c1_row.file1_segment2
      segment3 = c1_row.file1_segment3
      segment4 = c1_row.file1_segment4
  if highest_match_prob > threshold:
    df_.at[index, 'file3_id'] = row.file3_id
    df_.at[index, 'file3_description'] = row.file3_description
    df_.at[index, 'status'] = 'code1 solved'
    df_.at[index, 'file1_segment1'] = segment1
    df_.at[index, 'file1_segment2'] = segment2
    df_.at[index, 'file1_segment3'] = segment3
    df_.at[index, 'file1_segment4'] = segment4
    df_.at[index, 'file1_id'] = matching_id
    df_.at[index, 'file1_code'] = highest_match_code
    df_.at[index, 'file1_description'] = description
    df_.at[index, 'file1_probability'] = (highest_match_prob * 100)
  else:
    df_.at[index, 'file3_id'] = row.file3_id
    df_.at[index, 'file3_description'] = row.file3_description
    df_.at[index, 'status'] = 'code1 not solved'

df_
```

```
for index, row in df_p.iterrows():
  highest_match_prob = 0
  highest_match_code = ""
  matching_id = ""
  description = ""
  segment1 = ""
  segment2 = ""
  segment3 = ""
  segment4 = ""
  result = 0
  for c2_ind, c2_row in df_c2.iterrows():
    result = similar(row.file3_description, c2_row.file2_description)
    if result > highest_match_prob:
      highest_match_prob = result
      highest_match_code = c2_row.file2_code
      matching_id = c2_row.file2_id
      description = c2_row.file2_description
      segment1 = c2_row.file2_segment1
      segment2 = c2_row.file2_segment2
      segment3 = c2_row.file2_segment3
      segment4 = c2_row.file2_segment4
  if highest_match_prob > threshold:
    df_.at[index, 'file2_id'] = matching_id
    df_.at[index, 'file2_code'] = highest_match_code
    df_.at[index, 'status'] = 'code2 solved'
    df_.at[index, 'file2_description'] = description
    df_.at[index, 'file2_probability'] = (highest_match_prob * 100)
    df_.at[index, 'file2_segment1'] = segment1
    df_.at[index, 'file2_segment2'] = segment2
    df_.at[index, 'file2_segment3'] = segment3
    df_.at[index, 'file2_segment4'] = segment4
    
  else:
    df_.at[index, 'status'] = 'no solution'

df_
df_.to_csv('step1.csv')
files.download('step1.csv')
```
