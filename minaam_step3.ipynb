```
import pandas as pd
import io
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
from google.colab import files
uploaded = files.upload()

df_c1 = pd.read_csv(io.StringIO(uploaded['code1.csv'].decode('utf-8')))
col_names = []
for col in df_c1.columns:
  col_names.append('file1_' + col)
df_c1.columns = col_names
df_c1
```

```
from google.colab import files
uploaded = files.upload()
df_c2 = pd.read_csv(io.StringIO(uploaded['code2.csv'].decode('utf-8')))
col_names = []
for col in df_c2.columns:
  col_names.append('file2_' + col)
df_c2.columns = col_names
df_c2
```

```
from google.colab import files
uploaded = files.upload()
df_p = pd.read_csv(io.StringIO(uploaded['products.csv'].decode('cp1252')))
col_names = []
for col in df_p.columns:
  col_names.append('file3_' + col)
df_p.columns = col_names
df_p
```

```
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

df_ = pd.DataFrame(columns = ['file3_id', 'file3_description', 'file3_status'])
curr_index = 0
for index, row in df_p.iterrows():
  highest_match_prob1 = 0
  highest_match_index1 = 0
  highest_match_prob2 = 0
  highest_match_index2 = 0
  for c1_ind, c1_row in df_c1.iterrows():
    result = similar(row.file3_description, c1_row.file1_description)
    if highest_match_prob1 < result:
      highest_match_prob2 = highest_match_prob1
      highest_match_index2 = highest_match_index1
      highest_match_index1 = c1_ind
      highest_match_prob1 = result
  df_.at[curr_index, 'file3_id'] = df_p.at[index, 'file3_id']
  df_.at[curr_index, 'file3_description'] = df_p.at[index, 'file3_description']
  df_.at[curr_index, 'file3_status'] = 'processed dict1'
  df_.at[curr_index, 'file1_id'] = df_c1.at[highest_match_index1, 'file1_id']
  df_.at[curr_index, 'file1_segment1'] = df_c1.at[highest_match_index1, 'file1_segment1']
  df_.at[curr_index, 'file1_segment2'] = df_c1.at[highest_match_index1, 'file1_segment2']
  df_.at[curr_index, 'file1_segment3'] = df_c1.at[highest_match_index1, 'file1_segment3']
  df_.at[curr_index, 'file1_segment4'] = df_c1.at[highest_match_index1, 'file1_segment4']
  df_.at[curr_index, 'file1_description'] = df_c1.at[highest_match_index1, 'file1_description']
  df_.at[curr_index, 'file1_code'] = df_c1.at[highest_match_index1, 'file1_code']
  df_.at[curr_index, 'probability'] = highest_match_prob1
  curr_index += 1
  df_.at[curr_index, 'file3_id'] = df_p.at[index, 'file3_id']
  df_.at[curr_index, 'file3_description'] = df_p.at[index, 'file3_description']
  df_.at[curr_index, 'file3_status'] = 'processed dict1'
  df_.at[curr_index, 'file1_id'] = df_c1.at[highest_match_index1, 'file1_id']
  df_.at[curr_index, 'file1_segment1'] = df_c1.at[highest_match_index2, 'file1_segment1']
  df_.at[curr_index, 'file1_segment2'] = df_c1.at[highest_match_index2, 'file1_segment2']
  df_.at[curr_index, 'file1_segment3'] = df_c1.at[highest_match_index2, 'file1_segment3']
  df_.at[curr_index, 'file1_segment4'] = df_c1.at[highest_match_index2, 'file1_segment4']
  df_.at[curr_index, 'file1_description'] = df_c1.at[highest_match_index2, 'file1_description']
  df_.at[curr_index, 'file1_code'] = df_c1.at[highest_match_index2, 'file1_code']
  df_.at[curr_index, 'probability'] = highest_match_prob2
  curr_index += 1
  
for index, row in df_p.iterrows():
  highest_match_prob1 = 0
  highest_match_index1 = 0
  highest_match_prob2 = 0
  highest_match_index2 = 0
  for c2_ind, c2_row in df_c2.iterrows():
    result = similar(row.file3_description, c2_row.file2_description)
    if highest_match_prob1 < result:
      highest_match_prob2 = highest_match_prob1
      highest_match_index2 = highest_match_index1
      highest_match_index1 = c2_ind
      highest_match_prob1 = result
  df_.at[curr_index, 'file3_id'] = df_p.at[index, 'file3_id']
  df_.at[curr_index, 'file3_description'] = df_p.at[index, 'file3_description']
  df_.at[curr_index, 'file3_status'] = 'processed dict1'
  df_.at[curr_index, 'file2_id'] = df_c2.at[highest_match_index1, 'file2_id']
  df_.at[curr_index, 'file2_segment1'] = df_c2.at[highest_match_index1, 'file2_segment1']
  df_.at[curr_index, 'file2_segment2'] = df_c2.at[highest_match_index1, 'file2_segment2']
  df_.at[curr_index, 'file2_segment3'] = df_c2.at[highest_match_index1, 'file2_segment3']
  df_.at[curr_index, 'file2_segment4'] = df_c2.at[highest_match_index1, 'file2_segment4']
  df_.at[curr_index, 'file2_description'] = df_c2.at[highest_match_index1, 'file2_description']
  df_.at[curr_index, 'file2_code'] = df_c2.at[highest_match_index1, 'file2_code']
  df_.at[curr_index, 'probability'] = highest_match_prob1
  curr_index += 1
  df_.at[curr_index, 'file3_id'] = df_p.at[index, 'file3_id']
  df_.at[curr_index, 'file3_description'] = df_p.at[index, 'file3_description']
  df_.at[curr_index, 'file3_status'] = 'processed'
  df_.at[curr_index, 'file2_id'] = df_c2.at[highest_match_index1, 'file2_id']
  df_.at[curr_index, 'file2_segment1'] = df_c2.at[highest_match_index2, 'file2_segment1']
  df_.at[curr_index, 'file2_segment2'] = df_c2.at[highest_match_index2, 'file2_segment2']
  df_.at[curr_index, 'file2_segment3'] = df_c2.at[highest_match_index2, 'file2_segment3']
  df_.at[curr_index, 'file2_segment4'] = df_c2.at[highest_match_index2, 'file2_segment4']
  df_.at[curr_index, 'file2_description'] = df_c2.at[highest_match_index2, 'file2_description']
  df_.at[curr_index, 'file2_code'] = df_c2.at[highest_match_index2, 'file2_code']
  df_.at[curr_index, 'probability'] = highest_match_prob2
  curr_index += 1
  
df_
```
