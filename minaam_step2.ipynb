```
def get_jaccard_sim(str1, str2):
    str1 = str(str1)
    str2 = str(str2)
    a = set(str1.split()) 
    b = set(str2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```
def get_highest_match(rows1, rows2):
  highest_score = 0
  for row1 in rows1:
    for row2 in rows2:
      curr_score = get_jaccard_sim(row1, row2)
      if curr_score > highest_score:
        highest_score = curr_score
  return highest_score

df1_ = pd.DataFrame(columns = ['file1_var1', 'file1_var2', 'file1_var3', 'file1_var4', 'file1_var5'])
for ind, rows in df_c1.iterrows():
  c1_curr_highest = 0
  curr_index = ind
  for ind2, rows2 in df_c2.iterrows():
    c2_curr_highest = get_highest_match(rows, rows2)
    if c2_curr_highest > c1_curr_highest:
      df1_.at[ind, 'file1_var1'] = df_c1.at[ind, 'file1_var1']
      df1_.at[ind, 'file1_var2'] = df_c1.at[ind, 'file1_var2']
      df1_.at[ind, 'file1_var3'] = df_c1.at[ind, 'file1_var3']
      df1_.at[ind, 'file1_var4'] = df_c1.at[ind, 'file1_var4']
      df1_.at[ind, 'file1_var5'] = df_c1.at[ind, 'file1_var5']
      
      df1_.at[ind, 'file2_product_code'] = df_c2.at[ind2, 'file2_ProductCode']
      df1_.at[ind, 'file2_commodity'] = df_c2.at[ind2, 'file2_Commodity']
      df1_.at[ind, 'file2_phone'] = df_c2.at[ind2, 'file2_Phone']
      df1_.at[ind, 'file2_name'] = df_c2.at[ind2, 'file2_Name']
      df1_.at[ind, 'file2_gender'] = df_c2.at[ind2, 'file2_Gender']
      df1_.at[ind, 'file2_probability'] = str(c2_curr_highest)

df1_
```
